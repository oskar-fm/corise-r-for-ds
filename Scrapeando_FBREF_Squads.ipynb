```python
import pandas as pd
```

```python
df = pd.read_html("https://fbref.com/en/comps/12/stats/La-Liga-Stats", header=1)[0]
pd.set_option('display.max_columns', None)
df
```

```python
df.columns = [col.replace('.1', '/90') if col.endswith('.1') else col for col in df.columns]
df
```

```python
df['Dif_Goles_xG'] = df['Gls/90'] - df['xG/90']
df.sort_values(by='Dif_Goles_xG', ascending=False, inplace=True)
df.head()
```
